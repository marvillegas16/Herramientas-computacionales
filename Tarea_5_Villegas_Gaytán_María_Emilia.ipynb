**NUMPY**

1. Sumas de matrices rectangulares.

```python
import numpy as np

def suma_matrices(A, B):
    return A + B

A = np.array([[11, 7, -1], [-3, 5, 2], [4, 2, -2]])
B = np.array([[1, 1, 1], [10, 3, -5], [-6, -10, -3]])

print("Suma A + B:")
print(suma_matrices(A, B))
```

2. Multiplicacion entrada a entrada de matrices rectangulares

```python
def multiplicacion_elemento_por_elemento(A, B):
    return A * B

print("\nMultiplicación elemento por elemento A * B:")
print(multiplicacion_elemento_por_elemento(A, B))
```

 3. La traza de una matriz.

```python
def traza_matriz(M):
    return np.trace(M)

print("\nTraza de A:", traza_matriz(A))
print("Traza de B:", traza_matriz(B))
```

4. La transpuesta de una matriz.

```python
def transpuesta_matriz(M):
    return M.T

print("\nTranspuesta de A:")
print(transpuesta_matriz(A))
```

5. Encontrar el maximo y mınimo elemento en una matriz.

```python
def max_min_matriz(M):
    return np.max(M), np.min(M)

max_A, min_A = max_min_matriz(A)
print("\nMáximo de A:", max_A)
print("Mínimo de A:", min_A)
```

6. Calcular la suma del cuadrado de los elementos de una matriz.

```python
def suma_cuadrados_matriz(M):
    return np.sum(M**2)

print("\nSuma de cuadrados de A:", suma_cuadrados_matriz(A))
```

7. La potencia de matrices.

```python
def potencia_matriz(M, n):
    return np.linalg.matrix_power(M, n)

print("\nA al cuadrado:")
print(potencia_matriz(A, 2))
```

1. En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solucion del problema de
valor inicial:
y=−y + x y(−1) = 5. (1)
Los estudiantes llegaron a la solucion general cuya ecuacion es:
y = (x− 1) + Ce−x (2)
Apoye a sus compañeros y grafique en python la familia de curvas para la solucion de la ecuacion
diferencial y en un solo grafico muestre de diferente color las curvas para los siguientes tres casos:
Para valores de C > 0.
Cuando C = 0.
Para valores de C < 0.
Señale en el grafico las curvas que corresponden a cada uno de los tres casos.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_family_curves():
    x = np.linspace(-3, 3, 400)

    plt.figure(figsize=(10, 6))

    # C > 0 (azul)
    for C in [0.5, 1, 2, 3]:
        y = (x - 1) + C * np.exp(-x)
        plt.plot(x, y, 'b-', label=f'C = {C}' if C == 0.5 else "")

    # C = 0 (verde)
    y = x - 1
    plt.plot(x, y, 'g-', linewidth=2, label='C = 0')

    # C < 0 (rojo)
    for C in [-0.5, -1, -2, -3]:
        y = (x - 1) + C * np.exp(-x)
        plt.plot(x, y, 'r-', label=f'C = {C}' if C == -0.5 else "")

    # Solución particular con y(-1) = 5
    C_particular = (5 - (-1 - 1)) / np.exp(1)
    y_particular = (x - 1) + C_particular * np.exp(-x)
    plt.plot(x, y_particular, 'k--', linewidth=2, label=f'Solución particular (C ≈ {C_particular:.2f})')

    plt.title('Familia de soluciones para $y = (x - 1) + Ce^{-x}$')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_family_curves()
```

Utilizando la librerıa matplotlib:
Grafique la funcion con lınea continua para un rango de -10 a 10, con mas de 1000 muestras.
Investigue en la literatura como graficar las asıntotas horizontales y verticales. Muestrelas en el
grafico con lıneas punteadas.
Defina los lımites del eje y para evitar que la grafica muestre valores muy altos o muy bajos
cerca de las asıntotas verticales.
Anada una leyenda indicando los elementos presentes en la grafica.

```python
def plot_rational_function():
    x = np.linspace(-10, 10, 1000)
    y = (3*x + 7) / (x**2 - 4)

    # Eliminamos puntos cerca de las asíntotas verticales
    y[(x > -2.1) & (x < -1.9)] = np.nan
    y[(x > 1.9) & (x < 2.1)] = np.nan

    plt.figure(figsize=(10, 6))

    # Gráfica de la función
    plt.plot(x, y, 'b-', label='$f(x) = \\frac{3x + 7}{x^2 - 4}$')

    # Asíntotas verticales
    plt.axvline(x=2, color='r', linestyle='--', label='Asíntotas')
    plt.axvline(x=-2, color='r', linestyle='--')

    # Asíntota horizontal
    plt.axhline(y=0, color='g', linestyle='--', label='Asíntota horizontal')

    plt.title('Gráfica de función racional con asíntotas')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.ylim(-10, 10)  # Limitar el eje y para mejor visualización
    plt.legend()
    plt.grid(True)
    plt.show()

plot_rational_function()
```

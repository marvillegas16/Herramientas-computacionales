```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#from google.colab import files
#df=files.upload()
#print(df)

link='https://github.com/mike1502/HerramientasComputacionales/blob/main/Pandas/ebola_2014_2016_clean.csv'
code='?raw=true' #para que pandas lea el archivo csv y se pueda procesar
url=link+code # se unen el link y el c√≥digo para leer el archivo
df=pd.read_csv(url,sep=',',index_col=1) #cargar archivo a la variable df
```

```python
promedio_muertos=df['No. of confirmed deaths'].mean()
df['No. of confirmed deaths'].fillna(promedio_muertos,inplace=True)
```

```python
df.tail(10)
```

```python
df.isnull().sum() #repetimos el comando para verificar que ya no tenemos datos nulos
```

```python
#vamos a contar los casos por pa√≠s
conteo=df.groupby('Country').size().reset_index(name='conteo')
conteo=conteo.sort_values('conteo',ascending=False)
conteo
```

```python
data_missing=df['No. of confirmed deaths'].isnull().groupby(df['Country']).sum()
data_missing=data_missing.sort_values(ascending=False)
print(data_missing)
```

```python
plt.pie(conteo['conteo'],labels=conteo['Country'],autopct='%1.1f%%')
plt.axis('equal')
plt.title('Casos de Ebola por pa√≠s')
plt.savefig('CasosEbolaPorPais.png',dpi=300,format='png')
plt.show()
```

```python
guinea_infectados=Guinea['No. of confirmed cases'][0:-1:10] #asignamos variable de infectados
guinea_muertos=Guinea['No. of confirmed deaths'][0:-1:10] #asignamos variable para las muertes
fechas=Guinea.index[0:-1:10]  #extraemos las fechas para el rango propuesto [0:-1:10]
plt.figure(figsize=(10,5)) #comenzamos a graficar
plt.plot(fechas,guinea_infectados, 'o:r', ms=3, label='Infectados')
plt.plot(fechas,guinea_muertos,'o-b', ms=3,label='Muertos')
plt.grid() #malla de la gr√°fica
plt.title('Comparaci√≥n entre infectados y decesos')
plt.xlabel('Fechas')
plt.ylabel('N√∫mero de casos')
guinea_infectados=Guinea['No. of confirmed cases'][0:-1:10] #asignamos variable de infectados
guinea_muertos=Guinea['No. of confirmed deaths'][0:-1:10] #asignamos variable para las muertes
fechas=Guinea.index[0:-1:10]  #extraemos las fechas para el rango propuesto [0:-1:10]
plt.figure(figsize=(10,5)) #comenzamos a graficar
plt.plot(fechas,guinea_infectados, 'o:r', ms=3, label='Infectados')
plt.plot(fechas,guinea_muertos,'o-b', ms=3,label='Muertos')
plt.grid() #malla de la gr√°fica
plt.title('Comparaci√≥n entre infectados y decesos')
plt.xlabel('Fechas')
plt.ylabel('N√∫mero de casos')
plt.legend()
plt.xticks(rotation=90,fontsize=8)
plt.savefig('Infectadosmuertos.png', dpi=300, format='png')
plt.show()
```

```python
guinea_boys=Guinea['Boys'][0:-1:10]
guinea_girls=Guinea['Girls'][0:-1:10]
fechas=Guinea.index[0:-1:10]
plt.figure(figsize=(10,8))
plt.plot(fechas,guinea_boys,'o:r',ms=3, label='Ni√±os')
plt.plot(fechas,guinea_girls,'o-b',ms=3,label='Ni√±as')
plt.grid
plt.title('Comparacion entre n√±os y ni√±as')
plt.xlabel('Fechas')
plt.ylabel('Numero de casos')
plt.legend()
plt.xticks(rotation=90, fontsize=8)
plt.savefig('Ni√±osNi√±as.png', dpi=300, format='png')
plt.show()
```

```python
FC=Guinea['Female Cases'][0:-1:10]
G=Guinea['Girls'][0:-1:10]
AF=Guinea ['Adult Female'][0:-1:10]
OF=Guinea ['Old Female'][0:-1:10]

fechas=Guinea.index[0:-1:10]

plt.figure(figsize=(10,8))
plt.bar(fechas,FC, color='red', label='Total',alpha=0.6)
plt.bar(fechas,G, color='blue', label='Ni√±as', alpha=0.6)
plt.bar(fechas,AF,color='green', label='Adultas', alpha=0.6)
plt.bar(fechas, OF, color='yellow', label='Adultas mayores', alpha=0.6)
plt.grid()
plt.title('Casos por edad')
plt.xlabel('Fechas')
plt.ylabel('Numero de casos')
plt.legend()
plt.xticks(rotation=90, fontsize=8)
plt.savefig('Casos por edad.png', dpi=300, format='png')
plt.show()
```
